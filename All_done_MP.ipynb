```python
import cv2 
import time
import os
import sys
import tensorflow
from matplotlib import pyplot
from mtcnn.mtcnn import MTCNN
import tensorflow
import cvlib as cv

imagePath = sys.argv[1]  
capture   = cv2.VideoCapture(0) 


faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

recognizer  = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("koustubh1.yml")

capture.set(3, 640)
capture.set(4, 480)

img_counter = 0
frame_set   = []


init_time = time.time()

save2 = 'D:/All_img'
save  = 'D:/Cheat_img'

previous = time.time()
delta    = 0

imn=0
si =0

while True:
     
    ret,frame = capture.read()
    current   = time.time()
    delta    += current - previous
    previous  = current

    if (time.time() - init_time)>60:
        break

    if delta>=5:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        delta=0
        ky=55
        imn+=1
        faces = faceCascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5) #for face recognition
    
        #-------------------This part detects multifaces in the image--------------------------#  
       
        
        cv2.imwrite('multi_img.jpg',frame)
        
        filename = 'multi_img.jpg'
        
        pixels = pyplot.imread(filename)
        # create the detector, using default weights
        detector = MTCNN()
        # detect faces in the image
        faces2 = detector.detect_faces(pixels)
        cnt = 0
        for face in faces2:
            cnt+=1
            
      #------------------This part detects object in the image--------------------------------#
        
        im = cv2.imread('multi_img.jpg')
        bbox, label, conf = cv.detect_common_objects(im)
        
      #------------------This part detects face recognition-----------------------------------#
        
        
        for (x, y, w, h) in faces:
            roi_gray = gray[y:y+h, x:x+w] 
            id_, conf = recognizer.predict(roi_gray)
    
            if conf>=45 and conf <85:
                 ky=id_
                    
     #--------------------Image Classification------------------------------------------------#   
        
        if(cnt==0):
            si+=1
            cv2.imwrite(os.path.join(save2, str(imn)+"Zero_face.png"),frame)
        
        elif(cnt>1):
            si+=1
            cv2.imwrite(os.path.join(save2,str(imn)+"Multi_face.png"),frame)
            
        elif(ky!=0):
            si+=1
            cv2.imwrite(os.path.join(save2, str(imn)+"Not_koustubh.png"),frame)
        
        elif(len(label)>1):
            si+=1
            cv2.imwrite(os.path.join(save2, str(imn)+"Object_det.png"),frame)
            
        else:
            cv2.imwrite(os.path.join(save2, str(imn)+"clean.png"), frame)
        
        
        print("{} written!".format(img_counter))
        
    img_counter += 1
    
      #--------------------Percentage Calculation------------------------------------------------# 
        
percent=si*(100/imn)


print('******************************************************')
print('*                                                    *')
print('           Suspicious_Percentage is '+str(percent)+'%          ')
print('*                                                    *')
print('******************************************************')

     #---------------------Releasing Camera------------------------------------------------------#
capture.release()
cv2.destroyAllWindows()

```
